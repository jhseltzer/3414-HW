```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('virginia-history.csv', index_col=0)
num_positive_cases = np.flip(data.iloc[0:100]["positive"].to_numpy())

time_index = np.linspace(0,14,100)

fig = plt.figure(figsize=(8,6))
plt.plot(time_index,num_positive_cases,linewidth=2)
plt.grid()
plt.xlabel('$Time\ (Weeks)$',fontsize=16)
plt.ylabel('$Positive\ Cases$',fontsize=16)
plt.title('$Positive\ COVID-19\ Cases$',fontsize=18)
plt.show()
```

```python
newCasesPerDay = np.zeros(99)
prev = num_positive_cases[0]
for i in range(1,num_positive_cases.size):
    newCasesPerDay[i-1]=(num_positive_cases[i]-prev)
    prev = num_positive_cases[i]
```

```python
time_index = np.linspace(0,14,99)
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(time_index,newCasesPerDay,linewidth=2)
plt.grid()
plt.xlabel('$Time\ (Weeks)$',fontsize=16)
plt.ylabel('$Positive\ Cases$',fontsize=16)
plt.title('$New\ COVID-19\ Cases\ (First\ Week\ Not\ Included\ Because\ Previous\ Data\ DNE)$',fontsize=18)
plt.show()
```

```python
sum = 0
for i in newCasesPerDay:
    sum+=i
average_cases_per_day = sum/newCasesPerDay.size
```
